```python
import pandas as pd
from sklearn.cluster import MeanShift

# Load the data from the CSV file
data = pd.read_csv('uenaka_feature_cluster_issue.csv')

# Create a MeanShift object and fit the data to it
ms = MeanShift()
ms.fit(data)

# Obtain the cluster labels for each data point
labels = ms.labels_

# Print the unique labels and the number of data points in each cluster

ms_dataset = data.copy()
ms_dataset.loc[:,'Cluster'] = ms.labels_
ms_dataset.Cluster.to_frame()
```
